```python
train=pd.read_csv('../input/tabular-playground-series-jun-2021/train.csv')
test=pd.read_csv('../input/tabular-playground-series-jun-2021/test.csv')
# train.head()
test.head()
```

```python
train=train.drop(['id'],axis=1)
test=test.drop(['id'],axis=1)
```

```python
from sklearn.model_selection import train_test_split
X=train.drop(['target'],axis='columns',inplace=False)
y=train['target']
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.30,random_state=10)
```

```python
from xgboost import XGBClassifier
cld=XGBClassifier(random_state=1,use_label_encoder=True,n_estimators=90,learning_rate=0.04,verbosity=0)
cld.fit(X_train,y_train)
y_pred=cld.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,y_pred)*100)
```

```python
prediction=cld.predict_proba(X_test)
```

```python
sample_case=pd.read_csv('../input/tabular-playground-series-jun-2021/sample_submission.csv')
print(sample_case.head())
```

```python
y_output=cld.predict_proba(test)
submission=pd.DataFrame(y_output,columns=['Class_1','Class_2','Class_3','Class_4','Class_5','Class_6','Class_7','Class_8','Class_9'])
submission['id']=sample_case['id']
submission.to_csv('submission.csv',index=False)
submission
```
